# INTRODUCTION

```
import pandas as pd
url = "https://usafactsstatic.blob.core.windows.net/public/data/covid-19/covid_confirmed_usafacts.csv"
confirmed_df=pd.read_csv(url)
confirmed_df
```

After fetching the data, let's filter it just for NYC boroughs, remove some extraneous columns, and then transpose the data for graphing.

```
counties_df = confirmed_df.loc[(confirmed_df['State'] == 'NJ')]

counties_df=counties_df.set_index('County Name')
counties_df=counties_df.drop(columns=['countyFIPS', 'State', 'stateFIPS'])
counties_df= counties_df.iloc[:, :-1]

counties_df = counties_df.transpose()
counties_df = counties_df.rename_axis('Date')

counties_df.index =pd.to_datetime(counties_df.index)
counties_df
```

The first case was on Mar 5 in Bergen County.

Also, as NJ has 21 counties, we break up the visualization into subplots for North Jersey, Central Jersey and South Jersey using Wikipedia definitions.

```

df = counties_df
df = df.loc['2020-03-05':]

# we do this column reordering to ensure that the legend is listed in the desired order (i.e. Statewide, North Jersey, Central Jersey, South Jersey)
col = df.pop("Bergen County")
df.insert(1, col.name, col)

col = df.pop("Middlesex County")
df.insert(2, col.name, col)

northjersey_counties = ['Bergen County','Union County','Essex County','Hudson County','Morris County','Passaic County','Sussex County','Warren County']
centraljersey_counties = ['Middlesex County', 'Monmouth County', 'Ocean County', 'Mercer County','Somerset County', 'Hunterdon County']
southjersey_counties = ['Atlantic County', 'Burlington County','Camden County', 'Cape May County', 'Cumberland County', 'Gloucester County', 'Salem County']

df.insert(0, "Total", df.sum(axis=1))
df.insert(0, "South Jersey", df[southjersey_counties].sum(axis=1))
df.insert(0, "Central Jersey", df[centraljersey_counties].sum(axis=1))
df.insert(0, "North Jersey", df[northjersey_counties].sum(axis=1))

counties_df = df

df
```

Create the plotly graph for confirmed cases.

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def prepFigure(figure, title):
    figure.update_xaxes(tickformat='%a<br>%b %d',
        tick0 = '2020-03-22',
        dtick = 7 * 24 * 3600000,
        rangeselector=dict(
            buttons=list([
                dict(count=7, label="1w", step="day", stepmode="backward"),
                dict(count=14, label="2w", step="day", stepmode="backward"),
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=1, label="YTD", step="year", stepmode="todate"),
                dict(step="all")
            ])
        )
    )
    figure.update_layout(title="<b>" + title + "</b>", hovermode="x", height=1100)
    return figure

fig2 = make_subplots(rows=4, cols=1, 
                    shared_xaxes=True, 
                    vertical_spacing=0.06,
                    subplot_titles=("New Jersey","North Jersey", "Central Jersey", "South Jersey"))

for column in df:
  if column in northjersey_counties:
    fig2.add_scatter(x=df.index, y=df[column], mode='lines+markers', marker={"symbol": "cross"}, name=column, legendgroup="North Jersey", row=2, col=1)
  elif column in centraljersey_counties:
    fig2.add_scatter(x=df.index, y=df[column], mode='lines+markers', marker={"symbol": "square"}, name=column, legendgroup="Central Jersey", row=3, col=1)
  elif column in southjersey_counties:
    fig2.add_scatter(x=df.index, y=df[column], mode='lines+markers', marker={"symbol": "diamond"}, name=column, legendgroup="South Jersey", row=4, col=1)
  else:
    fig2.add_scatter(x=df.index, y=df[column], mode='lines', name=column, row=1, col=1, legendgroup="NJ")

prepFigure(fig2, "NJ COVID19 Confirmed Cases")
```

Now, let's graph the Daily incident counts from the cumulative data...

```
def dailydata(dfcounty):
    dfcountydaily=dfcounty.diff(axis=0).fillna(0)
    return dfcountydaily

DailyCases_df=dailydata(counties_df)
DailyDeaths_df=dailydata(counties_deaths_df)

# compute 7-day exponential moving average
DailyCases_df['EMA'] = DailyCases_df['Total'].ewm(span=7,adjust=False).mean()
DailyCases_df['North EMA'] = DailyCases_df['North Jersey'].ewm(span=7,adjust=False).mean()
DailyCases_df['Central EMA'] = DailyCases_df['Central Jersey'].ewm(span=7,adjust=False).mean()
DailyCases_df['South EMA'] = DailyCases_df['South Jersey'].ewm(span=7,adjust=False).mean()

DailyDeaths_df['EMA'] = DailyDeaths_df['Total'].ewm(span=7,adjust=False).mean()
DailyDeaths_df['North EMA'] = DailyDeaths_df['North Jersey'].ewm(span=7,adjust=False).mean()
DailyDeaths_df['Central EMA'] = DailyDeaths_df['Central Jersey'].ewm(span=7,adjust=False).mean()
DailyDeaths_df['South EMA'] = DailyDeaths_df['South Jersey'].ewm(span=7,adjust=False).mean()

DailyCases_df = DailyCases_df.loc['2020-03-08':]
DailyCases_df
```

```
df2 = DailyCases_df
fig2 = make_subplots(rows=4, cols=1, 
                    shared_xaxes=True, shared_yaxes=True,
                    vertical_spacing=0.06,
                    subplot_titles=("New Jersey","North Jersey", "Central Jersey", "South Jersey"))

fig2.add_scatter(x=df2.index, y=df2["Total"], mode='lines', name="<b>Statewide</b>", legendgroup="NJ", row=1, col=1)
fig2.add_scatter(x=df2.index, y=df2["North Jersey"], mode='lines', name="North Jersey",  line=dict(color="red"),legendgroup="NJ", row=1, col=1)
fig2.add_scatter(x=df2.index, y=df2["Central Jersey"], mode='lines', name="Central Jersey",  line=dict(color="green"),legendgroup="NJ", row=1, col=1)
fig2.add_scatter(x=df2.index, y=df2["South Jersey"], mode='lines', name="South Jersey",  line=dict(color="purple"),legendgroup="NJ", row=1, col=1)
fig2.add_scatter(x=df2.index, y=df2["Statewide Unallocated"], mode='lines', name="Statewide Unallocated", legendgroup="NJ", row=1, col=1)

fig2.add_scatter(x=df2.index, y=df2["North Jersey"], mode='lines', name="<b>North Jersey</b>", line=dict(color="red"), legendgroup="North Jersey", row=2, col=1)
fig2.add_scatter(x=df2.index, y=df2["Central Jersey"], mode='lines', name="<b>Central Jersey</b>", line=dict(color="green"), legendgroup="Central Jersey", row=3, col=1)
fig2.add_scatter(x=df2.index, y=df2["South Jersey"], mode='lines', name="<b>South Jersey</b>",  line=dict(color="purple"), legendgroup="South Jersey", row=4, col=1)

for column in df2:
  if column in northjersey_counties:
    fig2.add_scatter(x=df2.index, y=df2[column], mode='lines+markers', marker={"symbol": "cross"}, name=column, legendgroup="North Jersey", row=2, col=1)
  elif column in centraljersey_counties:
    fig2.add_scatter(x=df2.index, y=df2[column], mode='lines+markers', marker={"symbol": "square"}, name=column, legendgroup="Central Jersey", row=3, col=1)
  elif column in southjersey_counties:
    fig2.add_scatter(x=df2.index, y=df2[column], mode='lines+markers', marker={"symbol": "diamond"}, name=column, legendgroup="South Jersey", row=4, col=1)
  elif column == 'North EMA':
    fig2.add_scatter(x=df2.index, y=df2['North EMA'], mode='lines', line=dict(color="royalblue", width=4, dash='dot'), name='7 day EMA', legendgroup="North Jersey", row=2, col=1)
  elif column == 'Central EMA':
    fig2.add_scatter(x=df2.index, y=df2['Central EMA'], mode='lines', line=dict(color="royalblue", width=4, dash='dot'), name='7 day EMA', legendgroup="Central Jersey", row=3, col=1)
  elif column == 'South EMA':
    fig2.add_scatter(x=df2.index, y=df2['South EMA'], mode='lines', line=dict(color="royalblue", width=4, dash='dot'), name='7 day EMA', legendgroup="South Jersey", row=4, col=1)
  elif column == 'EMA':
    fig2.add_scatter(x=df2.index, y=df2['EMA'], mode='lines', line=dict(color="royalblue", width=4, dash='dot'), name='7 day EMA', legendgroup="NJ", row=1, col=1)

prepFigure(fig2, "NJ Daily New Cases")
#fig2.update_layout(title=, hovermode="x", height=670 )
```

```
url = "https://usafactsstatic.blob.core.windows.net/public/data/covid-19/covid_county_population_usafacts.csv"

county_population_df = pd.read_csv(url)

county_population_df

nj_counties_pop_df = county_population_df.loc[(county_population_df['State'] == 'NJ')]

nj_counties_pop_df=nj_counties_pop_df.drop(columns=['countyFIPS', 'State'])

nj_counties_pop_df=nj_counties_pop_df.set_index('County Name')

total= nj_counties_pop_df.sum()['population']
nj_counties_pop_df.loc['Total'] = [total]

subtotal = nj_counties_pop_df.loc[southjersey_counties].sum()['population']
nj_counties_pop_df.loc['South Jersey'] = [subtotal]

subtotal = nj_counties_pop_df.loc[centraljersey_counties].sum()['population']
nj_counties_pop_df.loc['Central Jersey'] = [subtotal]

subtotal = nj_counties_pop_df.loc[northjersey_counties].sum()['population']
nj_counties_pop_df.loc['North Jersey'] = [subtotal]

nj_counties_pop_df
```
